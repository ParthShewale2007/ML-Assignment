```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.

```python
years = ['2003', '2005', '2007', '2009', '2011']
sachin_runs = [1141, 1234, 1005, 915, 1356]
kohli_runs =  [0,    0,    325, 789, 1380]
sehwag_runs = [1250, 990, 1103, 1025, 987]

x = np.arange(len(years))
width = 0.25

plt.figure(figsize=(8, 5))
plt.bar(x - width, sachin_runs, width, label='Sachin')
plt.bar(x, sehwag_runs, width, label='Sehwag')
plt.bar(x + width, kohli_runs, width, label='Kohli')

plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs over Selected Years')
plt.xticks(x, years)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.Figure (figsize = (25,21))
plt.show()
```

### Horizontal Bar

```python
players = ['Sachin', 'Kohli', 'Sehwag']
total_runs_debut_5_years = [4500, 4800, 4200]

plt.figure(figsize=(8, 5))
plt.barh(players, total_runs_debut_5_years, color=['orange', 'blue', 'green'])
plt.xlabel('Total Runs')
plt.title('Total Runs in First 5 Years After Debut')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
